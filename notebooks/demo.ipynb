## Test Case 3 — Both Domains Together
Expected: Combined results from both domains showing 8 total emails.

```python
domains = ['testcorp.local', 'demo.org']
results = []
for domain in domains:
    r = harvest_domain(domain, verbose=True)
    print_result(r)
    results.append(r)
```

## Summary Table — All Test Cases

```python
all_reports = [r1, r2] + results
seen = set()
unique = []
for r in all_reports:
    if r['domain'] not in seen:
        seen.add(r['domain'])
        unique.append(r)

print(f"{'Domain':25s} {'Emails':>8}  {'Method':>18}  Timestamp")
print('-' * 75)
for r in unique:
    print(f"{r['domain']:25s} {len(r['emails']):>8}  "
          f"{r['method']:>18}  {r['timestamp']}")
print('-' * 75)
total = sum(len(r['emails']) for r in unique)
print(f"Total emails harvested across all domains: {total}")
```
